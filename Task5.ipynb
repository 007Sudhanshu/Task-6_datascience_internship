```python
#Q1. How many people have heart disease?
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=("orchid","salmon"))
plt.title("Heart Disease")
plt.xlabel("1=Yes ,0=No")
plt.ylabel("Amount")
plt.show()
```

```python
df.target.value_counts().plot(kind='pie',figsize = (10,8))
plt.legend("disease","no disease")
plt.show()
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie',figsize = (10,8))
plt.legend("Males","Females")
plt.show()
```

```python
#Q2. How of which sex have most heart disease?
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x = 'target', data = df, hue = 'sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0 = No heart Disease, 1 Heart Disease");
```

```python
#Q3.People of which sex has which type of chest pain?
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind= 'bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title("Chest pain type vs pain");
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind= 'bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title("Type of chest pain for sex")
plt.xlabel("0= Female, 1= Male")
```

```python
#Q4. Which chest pain are most prone to heart disease?
pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x = 'cp', data = df, hue = 'target')
plt.title("Pain most prone to heart attack")
```
