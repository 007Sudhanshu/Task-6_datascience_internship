```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("householdtask3.csv")
```

```python
data.head(10)
```

```python
plt.scatter(data['year'],data['own'])
plt.title("Scatter plot")
plt.xlabel("Year")
plt.ylabel("Own")
plt.show()
```

```python
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("Line plot")
plt.xlabel("Year")
plt.ylabel("own")
plt.show()
```

```python
plt.bar(data['year'],data['own'])
plt.title("Scatter plot")
plt.xlabel("Year")
plt.ylabel("Own")
plt.show()
```

```python
plt.hist(data['income'])
plt.title("Histogram")
plt.show()
```
